```python
from openpyxl import load_workbook
from datetime import datetime, timedelta
import pandas as pd 
import numpy as np
```

```python
def parse_trash(all_sheets, wb):
    data=[]
    total=[]
    by_day=[]
    data_local= []
    for j in all_sheets:
        sheet = wb.get_sheet_by_name(j)
        if (j[2]=='-'):
            j=j[3:len(j)]
            date_l=datetime.strptime(j, "%d.%m.%y")
        #print(date_l.year, date_l.month, date_l.day)
            for i in range(20, 23):
                date_e=datetime.strptime(str(sheet.cell(row=i, column=34).value), "%H:%M:%S")
                total.append(sheet.cell(row=i, column=35).value)
                by_day.append(sheet.cell(row=i, column=36).value)
                #print(date_e.hour, date_e.minute, date_e.second)
                data.append(datetime(date_l.year, date_l.month, date_l.day,date_e.hour, date_e.minute, date_e.second)-timedelta(days=1))
            for i in range(23, 32):
                date_e=datetime.strptime(str(sheet.cell(row=i, column=34).value), "%Y-%m-%d %H:%M:%S")
                data.append(datetime(date_l.year, date_l.month, date_l.day,date_e.hour, date_e.minute, date_e.second))
                total.append(sheet.cell(row=i, column=35).value)
                by_day.append(sheet.cell(row=i, column=36).value)
        else:
            date_l=datetime.strptime(j, "%d.%m.%y")
            for i in range(20, 32):
                date_e=datetime.strptime(str(sheet.cell(row=i, column=34).value), "%H:%M:%S")
                data.append(datetime(date_l.year, date_l.month, date_l.day,date_e.hour, date_e.minute, date_e.second))
                total.append(sheet.cell(row=i, column=35).value)
                by_day.append(sheet.cell(row=i, column=36).value)


    d = {'Time': data, 'Vsego': total, 'Za chas': by_day}
    dp=pd.DataFrame(data=d)
    return dp
```

```python
# Load in the workbook
wb = load_workbook('October_1.xlsm',  data_only=True)

# Get sheet names
print(wb.sheetnames)
```
